# Step 3: Clean up and add data to prepare for analysis

Inspect the new table that has been created.

```r
colnames(all_trips)
nrow(all_trips)
dim(all_trips)
head(all_trips)
str(all_trips)
summary(all_trips)
```

Removing some columns from the dataset.

```r
all_trips$start_lat <- NULL
all_trips$start_lng <- NULL
all_trips$end_lat <- NULL
all_trips$end_lng <- NULL
all_trips$weekday <- NULL
all_trips$from_station_name <- NULL
all_trips$to_station_name <- NULL
all_trips$from_station_id <- NULL
all_trips$to_station_id <- NULL
```

There are a few problems we will need to fix:
1. The data can only be aggregated at the ride level, which is too granular. We will want to add some additional columns to the data -- such as Day, Month, Year -- that provides additional opportunities to aggregate the data.
2. We will want to add a calculated field for length of ride ("ride_length").
3. There are some rides where ride_length is NA, including several hundred rides where Divvy took bikes out of circulation for Quality Control reasons. We will want to delete these rides.

Add columns that list the date, month, day and year of each ride.
This will allow us to aggregate ride data for each month, day or year.

```r
all_trips$starting_date <- as.Date(all_trips$starting_date, "%d/%m/%Y")
all_trips$date <- as.Date(all_trips$starting_date)
all_trips$month <- format(as.Date(all_trips$date), "%m")
all_trips$day <- format(as.Date(all_trips$date),"%d")
all_trips$year <- format(as.Date(all_trips$date), "%Y")
all_trips$Day_of_week <- format(as.Date(all_trips$date),"%A")
```

Add a "ride_length" calculation to all_trips(in seconds).

```r
all_trips$ride_length <- strptime(all_trips$ride_duration, format = "%H:%M:%S")

all_trips$ride_length <- hour(all_trips$ride_length) * 3600 + minute(all_trips$ride_length) * 60 + second(all_trips$ride_length)
```

Remove bad data.
The dataframe includes few hundred entries where bikes are taken out of docks and check for quality by Divvy is NA.
We will create new version of the dataframe (v2) since the data is being removed.

```r
all_trips_v2 <- na.omit(all_trips)
```

# Step 4: Conduct descriptive analysis

Descriptive analysis on ride length (all figures are in seconds).

```r
summary(all_trips_v2)
```

```r
summary(all_trips_v2$ride_length)
```

Compare member and casual user.

```r
aggregate(all_trips_v2$ride_length ~ all_trips_v2$usertype, FUN = mean)
aggregate(all_trips_v2$ride_length ~ all_trips_v2$usertype, FUN = median)
aggregate(all_trips_v2$ride_length ~ all_trips_v2$usertype, FUN = max)
aggregate(all_trips_v2$ride_length ~ all_trips_v2$usertype, FUN = min)
```

See the average ride time by *each day* for member vs casual users

```r
aggregate(all_trips_v2$ride_length ~ all_trips_v2$usertype + all_trips_v2$Day_of_week, FUN = mean)
```

Notice that the Days_of_week are out of order. Let's fix that.

```r
all_trips_v2$Day_of_week <- ordered(all_trips_v2$Day_of_week, levels=c("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"))
```

Now, let's run the average ride time by each day for member vs casual users.


```r
aggregate(all_trips_v2$ride_length ~ all_trips_v2$usertype + all_trips_v2$Day_of_week, FUN = mean)
```

See the average ride time by *month* for member and casual users.

```r
#put months in order

all_trips_v2$month <- ordered(all_trips_v2$month, levels = c("01", "02", "03", "04", "05", "06","07", "08", "09", "10", "11", "12"))

#average ride time by month
aggregate(all_trips_v2$ride_length ~ all_trips_v2$usertype + all_trips_v2$month, FUN =  mean)
```

```r
# Number of rides between members and casual users for each month

all_trips_v2 %>% 
  group_by(usertype, month) %>% 
  summarise(number_of_rides = n(), .groups = 'drop') %>% 
  arrange(month)
```

See the bike type preference between casual and member users.

```r
#compairing bike type preference between casual and member users

all_trips_v2 %>% 
  group_by(rideable_type, usertype) %>% 
  summarise(number_of_rides = n(), .groups = 'drop')
```

Analyze ridership data by usertype and Day_of_week.

```r
all_trips_v2 %>% 
  group_by(usertype,Day_of_week) %>% #groups by usertype and Day_of_week
  summarise(number_of_rides = n(), average_duration = mean(ride_length), .groups = 'drop') %>%  # calculate the number of rides and average duration
  arrange(usertype, Day_of_week) #sorts
```

# Creating Data Visualizations

```r
my_theme = theme(plot.title=element_text(size=20),
                 axis.text.x=element_text(size=16), 
                 axis.text.y=element_text(size=16),
                 axis.title.x=element_text(size=18), 
                 axis.title.y=element_text(size=18),
                 strip.text.x=element_text(size=16), 
                 strip.text.y=element_text(size=16),
                 legend.title=element_text(size=18), 
                 legend.text=element_text(size=16))
```

```r
options(scipen = 999) #to disable scientific notation in plot.
```

**Viz 1 -** Let's visualize the number of rides by usertype.

```r
options(repr.plot.width = 10, repr.plot.height = 8)

all_trips_v2 %>% 
  group_by(usertype, Day_of_week) %>% 
  summarise(number_of_rides = n(), average_duration = mean(ride_length), .groups ='drop') %>%  
  ggplot(aes(x = Day_of_week, y = number_of_rides, fill = usertype)) +
  geom_col(position = "dodge") + labs(title = "Number of rides by day: Member vs Casual users ") + my_theme
  
```

**Viz 2-** Let's create a visualization for average duration.

```r

options(repr.plot.width = 10, repr.plot.height = 8)

all_trips_v2 %>% 
  group_by(usertype, Day_of_week) %>% 
  summarise(average_duration = mean(ride_length), .groups = 'drop') %>%  
  ggplot(aes(x = Day_of_week, y = average_duration, fill = usertype)) +
  geom_col(position = "dodge") + labs(title = "Average ride duration by day: Member vs  Casual users") +
my_theme

```

**Viz 3-** Let's create visualization for different bike usage by member and casual users.

```r
options(repr.plot.width = 12, repr.plot.height = 8)

all_trips_v2 %>% 
  group_by(rideable_type, usertype) %>% 
  summarise(number_of_rides = n(), .groups = 'drop') %>% 
  ggplot(aes(x = usertype, y = number_of_rides, fill = usertype)) +
  geom_col(position = "dodge") +
  facet_wrap(~rideable_type) +
  labs(x = " ", title = "Usage of Different Bikes: Members vs. Casual users") + my_theme
```
