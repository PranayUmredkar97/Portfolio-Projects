# Cleaning and preparing data for analysis

```python
#Dropping some columns as it is not reuired for analysis.
match_df.drop(['umpire1', 'umpire2'], axis=1, inplace=True)
match_df.head(2)
```

```python
ball_df.drop(['fielder'],axis=1, inplace=True)
ball_df.head(2)
```

```python
match_df['toss_winner'].value_counts()
```

There are some typos in team name of Rising pune supergiant so we are going to replace the name to make it consistent.
There are also some team name which has been changed by their organization such as earlier Delhi capitals name was Delhi Daredevils but we are not changing them.

```python
# as we can see there are error in data we are seeing two teams having same name with some typos
# we will replace name to make it consistent
match_df['toss_winner'].replace(['Rising Pune Supergiants'], 'Rising Pune Supergiant', inplace =True)
match_df['toss_winner'].value_counts()
```

```python
match_df['city'].value_counts()
```

```python
#There are two names for one city Bangalore and Bengaluru so we are replacing them to one to make it consistent.
match_df['city'].replace(['Bangalore'],'Bengaluru', inplace=True)
match_df['city'].value_counts()
```

```python
# extracting the year from date column and storing it into new column.
match_df['season'] = pd.DatetimeIndex(match_df['date']).year
match_df.head(5)
```

#### Total matches played in each season.

```python
match_per_season = match_df.groupby(['season'])['id'].count().reset_index().rename(columns=({'id': 'matches'}))
match_per_season
```

```python
#plotting bar graph for total number of matches played in each season.
sns.set_style(style = 'darkgrid')
plt.figure(figsize= (12,8))
sns.countplot(x = 'season',  data = match_df)
plt.xticks(rotation = 45, fontsize = 16)
plt.yticks(fontsize = 16)
plt.xlabel('Season', fontsize = 20)
plt.ylabel('Count', fontsize = 20)
plt.title('Matches played in each season', fontsize = 22, fontweight = 'bold');
```

#### Merging columns using left  join.

```python
season_data = match_df[['id', 'season']].merge(ball_df, left_on= 'id', right_on= 'id', how ='left').drop('id', axis = 1)
season_data.head(5)
```

#### Total runs scroed in each season.

```python
season = season_data.groupby(['season'])['total_runs'].sum().reset_index()
season
```

```python
#Plotting linechart for total runs scored in each season
season = season_data.groupby(['season'])['total_runs'].sum().reset_index()
plt.figure(figsize=(12,8))
sns.lineplot(x='season', y='total_runs', data = season)
plt.xlabel('Season', fontsize = 20)
plt.ylabel('Runs', fontsize = 20)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.title('Total runs scored in each season', fontsize= 22, fontweight = 'bold');
```

#### Number of tosses won by each team.

```python
toss = match_df['toss_winner'].value_counts()
toss
```

```python
#plotting number of toss won by each team  
plt.figure(figsize=(12,8))
sns.barplot(x = toss, y =toss.index)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.xlabel('Number of tosses won', fontsize=20)
plt.ylabel('Teams', fontsize=20)
plt.title("Number of tosses won by each team", fontsize=22, fontweight='bold');
```

```python
#Let's look at the runs scored by MS Dhoni from single, double, triple, fours and six in IPL.
def count(df_MSD, runs):
    return len(df_MSD[df_MSD['batsman_runs']==runs])*runs
```

```python
print("Runs scored from 1's :", count(df_MSD, 1))
print("Runs scored from 2's :", count(df_MSD, 2))
print("Runs scored from 3's :", count(df_MSD, 3))
print("Runs scored from 4's :", count(df_MSD, 4))
print("Runs scored from 5's :", count(df_MSD, 6))
```

```python
#Let's see top 10 bastman with highest runs scored in IPL.
runs = ball_df.groupby(['batsman'])['batsman_runs'].sum().reset_index()
runs.columns = ['batsman', 'runs']
y = runs.sort_values(by='runs', ascending=False).head(10).reset_index().drop('index', axis=1)
y
```

```python
#Visualization of top 10 run scorer in IPL.
plt.figure(figsize=(12,8))
sns.barplot(x=y['batsman'],y=y['runs'], palette='rocket')
plt.xticks(fontsize=16, rotation=45)
plt.yticks(fontsize=16)
plt.xlabel('Player',fontsize=20)
plt.ylabel('Runs',fontsize=20)
plt.title('Top 10 runs scorer in IPL', fontsize=22, fontweight='bold');
```

# Inferences and Conclusion

* From season 2008 until season 2020, a total of 816 IPL matches have been played.
* The most matches were played in seasons 2011, 2012, and 2013 because there were more teams participating.
* The total number of runs scored in each season is highest in 2012 (22541), followed by 2012 and 2011, and lowest in 2009(16320).
* Mumbai Indians (106) won the most tosses, followed by Kolkata Knight Riders (98) and Chennai Super Kings (97).
* Except for seasons 2009, 2010, and 2013, the most preferred decision was fielding after winning the toss.
* The majority of matches are won by chasing runs.
* The most matches were played at Eden Gardens Stadium (77), followed by Feroj Shah Kotla and Wankhede Stadium.
* The Mumbai City hosted the most matches (101) followed by Bengaluru and Kolkata.
* The best stadium for defending runs is Eden Gardens, while the best stadium for chasing is Feroj Shah Kotla.
* The Kolkata Knight Riders (58) and Mumbai Indians (58) won more games chasing runs, while the Mumbai Indians (64) and Chennai Super Kings (53) won more games defending runs.
* AB de Villiers (23) has the most Player of the Match awards , followed by Chris Gayle (22) and Rohit Sharma (18).
* Mumbai Indians have won the most IPL titles (5), followed by Chennai Super Kings (3).
* Virat Kohli has the most IPL runs (5878), followed by Suresh Raina (5368) and David Warner (5254).
